Flowchart- Simple Interest
Start
Input Principal(P), Rate(R), Time(T)
Compute Amount(A)= P*(1+(R/100)*T)
Display A
End


```python
#SImple Interest
P= float(input("Enter Principal:"))
R= float(input("Enter Rate(%):"))
T= float(input("Enter Time (years):"))

A= P*(1+ (R/100)*T)

print("Amount =",A)
```

FLowchart for Compound Interest
Start
Input P, R, n, t
Compute A = P*(1+R/n)^(n*t)
Display A
End

```python
#Compound Interest
P= float(input("Enter Principal:"))
R= float(input("Enter Rate(%):"))/100
n= int(input("Enter number of times compounded per year:"))
t= float(input("Enter time in years:"))

A= P*(1+ R/n)**(n*t)

print("Amount =",A)
```

Flowchart for Annuity Plan
Start-
Input PMT, R, n, t-
Compute A=PMT*((1+R/n)^(n*t)-1)/(R/n)-
Display A-
End

```python
#Annuity Plan
PMT= float(input("Enter periodic payment (PMT):"))
R= float(input("Enter Rate (%):"))/100
n= int(input("Enter number of payments per year:"))
t= float(input("Enter number of years:"))

A= PMT* ((1+R/n)**(n*t)-1)/(R/n)
print("Future Value of Annuity =",A)
```
