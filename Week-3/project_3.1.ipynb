```python
import tkinter as tk
from tkinter import messagebox
import csv

#Load CSV data
employees=[]
with open("GIG-logistics.csv","r")as file:
    reader=csv.DictReader(file)
    for row in reader:
        employees.append(row)

#Verify employee function
def verify_employee():
    name=name_entry.get().strip().lower()
    dept= dept_entry.get().strip().lower()

    found= False
    members=[]

    for emp in employees:
        if emp["Name"].lower()==name and emp["Department"].lower()==dept:
            found= True

        if emp["Department"].lower()==dept and emp["Name"].lower()!=name:
            members.append(emp["Name"])

        if found:
            result=f"Welcome {name.title()}!\n\nOther members in {dept.title()} department:\n"
            result+="\n".join(members)
            output_label.config(text=result)
        else:
            output_label.config(text="Employee does not exist")

#Main window
window=tk.Tk()
window.title("GIG Logistics Employee Verification")
window.geometry("420x300")
window.configure(bg="white")

#Title
title=tk.Label(window, text="Employee Verification System",
               font=("Arial",16,"bold"),bg="white")
title.pack(pady=10)

#Frame for inputs
frame=tk.Frame(window, bg="white")
frame.pack(pady=10)

#Name
tk.Label(frame, text="Employee Name:", bg="white").grid(row=0,column=0,padx=10,pady=5)
name_entry=tk.Entry(frame,width=25)
name_entry.grid(row=0,column=1,pady=5)

#Department
tk.Label(frame,text="Department:", bg="white").grid(row=1,column=0,padx=10,pady=5)
dept_entry=tk.Entry(frame, width=25)
dept_entry.grid(row=1,column=1,pady=5)

verify_btn=tk.Button(window, text="Verify Employee",
                           command=verify_employee,
                           bg="blue", fg="white", width=20)
verify_btn.pack(pady=15)

#Output
output_label=tk.Label(window, text="",bg="white",justify="left")
output_label.pack(pady=10)

window.mainloop()
```
