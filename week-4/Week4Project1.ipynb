```python
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/HP/Downloads/archive/AB_NYC_2019.csv")
data
```

```python
print(data.head())
print(data.info())
```

```python
#Remove duplicates
data = data.drop_duplicates()
#Handle missing values
data = data.dropna(subset=['name','host_name'])
#Fill missing reviews with 0
data['reviews_per_month']=data['reviews_per_month'].fillna(0)
print(data.isnull().sum())
```

```python
data= data.drop('last_review',axis=1)
print(data.isnull().sum())
```

```python
print(data.info())
```

```python
import matplotlib.pyplot as plt
plt.hist(data['price'],bins=50)
plt.title("Price Distribution")
plt.show()
```

```python
data['room_type'].value_counts().plot(kind='bar')
plt.title("Room Type Distribution")
plt.show()
```

```python
data['neighbourhood_group'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Listings by Area")
plt.show()
```

INSIGHTS
1. Most listings are low-priced.
2. Few very expensive outliers.
3. Entire homes may dominate.
4. Private rooms are also common.
5. Manhattan usually has the highest listings.
6. Staten Island has the least listings.
