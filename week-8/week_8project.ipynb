```python
import tkinter as tk
from tkinter import messagebox

# Parent Class
class Zenith:
    def mutual_services(self):
        return ["Lines of credit", "Investment management and accounts", "Insurance"]

    def unique_services(self):
        return []

# Child Classes
class RetailBanking(Zenith):
    def unique_services(self):  # Polymorphism (override)
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]

class GlobalBanking(Zenith):
    def unique_services(self):  # Polymorphism (override)
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]

class CommercialBanking(Zenith):
    def unique_services(self):  # Polymorphism (override)
        return [
            "Advisory services"
        ]

# GUI Function
def show_services():
    name = name_entry.get()
    division = division_var.get()

    if not name or not division:
        messagebox.showwarning("Input Error", "Please enter name and select division")
        return

    # Object creation based on division
    if division == "Retail":
        obj = RetailBanking()
    elif division == "Global":
        obj = GlobalBanking()
    elif division == "Commercial":
        obj = CommercialBanking()
    else:
        return

    # Get services
    mutual = obj.mutual_services()
    unique = obj.unique_services()

    # Display result
    result = f"Employee: {name}\nDivision: {division}\n\n"
    result += "Mutual Services:\n" + "\n".join(mutual) + "\n\n"
    result += "Unique Services:\n" + "\n".join(unique)

    output_label.config(text=result)

# GUI Setup
root = tk.Tk()
root.title("Zenith Bank Services App")
root.geometry("500x400")

# Name Input
tk.Label(root, text="Employee Name:").pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Division Selection
tk.Label(root, text="Select Division:").pack()
division_var = tk.StringVar()

tk.Radiobutton(root, text="Retail Banking", variable=division_var, value="Retail").pack()
tk.Radiobutton(root, text="Global Banking", variable=division_var, value="Global").pack()
tk.Radiobutton(root, text="Commercial Banking", variable=division_var, value="Commercial").pack()

# Button
tk.Button(root, text="Show Services", command=show_services).pack(pady=10)

# Output
output_label = tk.Label(root, text="", justify="left")
output_label.pack()

root.mainloop()
```
