Pseudocode for Store Checkout
START
SET Total=0
INPUT Numberofitems
FOR Item=1 TO Numberofitems
INPUT Price
Total=Total+Price
END FOR

OUTPUT "Total Amount=", Total

INPUT AmountPaid
IF AmountPaid>=Total THEN
Change=AmountPaid-Total
OUTPUT "Payment Successful"
ELSE
OUTPUT "Insufficient Payment"
END IF

END

PSEUDOCODE that takes as input two names and their corresponding ages and then swap the ages.

INPUT Name1
INPUT Age1
INPUT Name2
INPUT Age2
Temp=Age1
Age1=Age2
Age2=Temp
OUTPUT Name1, Age1
OUTPUT Name2, Age2
END
